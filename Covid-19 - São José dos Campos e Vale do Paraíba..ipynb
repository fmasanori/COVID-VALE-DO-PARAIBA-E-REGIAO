```python
import pandas as pd
low_memory = False
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
#Import de bibliotecas 
```

```python
filename = r"C:\Users\Kevin Melo\Desktop\Masanori\HIST_PAINEL_COVIDBR_02jul2020.xlsx"
df = pd.read_excel(filename)
#Abrindo o excel da base de dados
```

```python
df.columns
#Vendo as colunas do excel
```

```python
df.rename(columns={'municipio' : 'cidade', 'semanaEpi': 'semana', 'nomeRegiaoSaude' : 'nomeRegiao'},inplace=True)
#Renomeando as colunas que são necessárias
```

```python
avp=df.query('nomeRegiao == "ALTO VALE DO PARAIBA"')  #nomendo a váriavel avp = ALTO VALE DO PARAIBA 
vp=df.query('nomeRegiao == "VALE DO PARAIBA/REGIAO SERRANA"') #nomendo a váriavel vp = VALE DO PARAIBA/REGIAO SERRANA
avp.groupby('cidade')['casosAcumulado'].max().sort_values().tail(5).plot(kind='pie',figsize=(8,8))  #Criando o gráfico de casos acumulados em formato de pizza nas cidades da avp
```

```python
vp.groupby('cidade')['casosAcumulado'].max().sort_values().tail(5).plot(kind='pie',figsize=(8,8))
#Criando o gráfico de casos acumulados em formato de pizza nas cidades da vp
```

```python
avp.groupby('cidade')['casosNovos'].max().sort_values().tail(15).plot(kind='bar',figsize=(8,8))
#Criando o gráfico de casos novos em formato de barra nas cidades da avp
```

```python
vp.groupby('cidade')['casosNovos'].max().sort_values().tail(15).plot(kind='bar',figsize=(9,9))
#Criando o gráfico de casos novos em formato de barra nas cidades da vp
```

```python
avp.groupby('semana')['obitosAcumulado'].max().tail(15).plot(kind='barh',figsize=(8,8))
#Criando o gráfico de óbitos acumulados em formato de barra horizontal por semanas da avp
```

```python
vp.groupby('semana')['obitosAcumulado'].max().tail(15).plot(kind='barh',figsize=(8,8))
#Criando o gráfico de óbitos acumulados em formato de barra horizontal por semanas da vp
```

```python
avp.groupby('semana')['obitosNovos'].max().tail(15).plot(kind='barh',figsize=(8,8))
#Criando o gráfico de óbitos novos em formato de barra horizontal por semanas da avp
```
